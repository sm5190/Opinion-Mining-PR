```python
import snscrape.modules.twitter as sntwitter
import pandas as pd

# Creating list to append tweet data to
tweets_list2 = []

# Using TwitterSearchScraper to scrape data and append tweets to list
for i,tweet in enumerate(sntwitter.TwitterSearchScraper('Palestine near:"India" since:2021-05-01 until:2021-05-19').get_items()):
    tweets_list2.append([tweet.date, tweet.content])
    
# Creating a dataframe from the tweets list above
tweets_df2 = pd.DataFrame(tweets_list2, columns=['Datetime', 'Text'])
tweets_df2.to_csv('Palestine_Israel.csv')
```

```python
import snscrape.modules.twitter as sntwitter
import pandas as pd

# Creating list to append tweet data to
tweets_list2 = []

# Using TwitterSearchScraper to scrape data and append tweets to list
for i,tweet in enumerate(sntwitter.TwitterSearchScraper('Pakistan near:"India" since:2015-05-01 until:2021-05-19').get_items()):
    tweets_list2.append([tweet.date, tweet.content])
    
# Creating a dataframe from the tweets list above
tweets_df2 = pd.DataFrame(tweets_list2, columns=['Datetime', 'Text'])
tweets_df2 = tweets_df2[tweets_df2['lang']=='en']
tweets_df2.to_csv('Pakistan_india.csv')
```
